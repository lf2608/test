```python
def get_coordinate(places):
    gmaps = googlemaps.Client(key=YOUR_API_KEY)
    coor = defaultdict(dict)
    for i,p in enumerate(places):
        p = str(p)
        geocode_result = gmaps.geocode(p)
        x,y = geocode_result[0]['geometry']['location']['lat'],geocode_result[0]['geometry']['location']['lng']
        add = geocode_result[0]['formatted_address']
        coor[p]['address'] = add
        coor[p]['coor'] = (x,y)
        #return coor
        return(coor)
```

```python
def get_coordinate_pd(places):
    gmaps = googlemaps.Client(key=YOUR_API_KEY)
    coor = defaultdict(list)
    num = len(places)
    count = 0
    for i,p in tqdm_notebook(enumerate(places)):
        #print(p)
        count += 1
        #p = str(p)
        geocode_result = gmaps.geocode(p)
        try:
            x,y = geocode_result[0]['geometry']['location']['lat'],geocode_result[0]['geometry']['location']['lng']
            add = geocode_result[0]['formatted_address']
            coor['Hospital'].append(p)
            coor['Address'].append(add)
            coor['Coordinates'].append((x,y))
        except:
            print('error',p)
            coor['Hospital'].append(p)
            coor['Address'].append('nan')
            coor['Coordinates'].append('nan')
        if count == 100:
            time.sleep(float(np.random.rand(1)))
    return coor
```
